## Question 6. Median value of horsepower

 1. Find the median value of the horsepower column in the dataset.
 2. Next, calculate the most frequent value of the same horsepower column.
 3. Use the fillna method to fill the missing values in the horsepower column with the most frequent value from the previous step.
 4. Now, calculate the median value of horsepower once again.

Has it changed?

```python
# Find the median value of the horsepower column
median = df['horsepower'].median()
median
```

```python
# Find the most frequent value in the horsepower column
mode = df['horsepower'].mode()[0]
mode
```

```python
# Use the fillna method to fill the missing values in the horsepower column with the most frequent value from the previous step
df['horsepower'].fillna(mode, inplace=True)
```

```python
# Alternatively
df['horsepower'] = df['horsepower'].fillna(mode)
```

```python
df.isnull().sum()
```

```python
# Check if the value of median changed
df['horsepower'].median()
```

Yes, the median changed-- it increased.

## Question 7. Sum of weights

1. Select all the cars from Asia
2. Select only columns vehicle_weight and model_year
3. Select the first 7 values
4. Get the underlying NumPy array. Let's call it X.
5. Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
6. Invert XTX.
7. Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200].
8. Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
9. What's the sum of all the elements of the result?

```python
# Select all the cars from Asia
df_asia = df[df['origin'] == 'Asia']
df_asia
```

```python
# Select only columns vehicle_weight and model_year
df_filtered = df_asia[['vehicle_weight', 'model_year']]
df_filtered
```

```python
# Select the first 7 values
df_head = df_filtered.head(7)
df_head
```

```python
# Get the underlying NumPy array. Let's call it X.
X = np.array(df_head)
X
```

```python
# Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
X_T = X.T
X_T
```

```python
XTX = np.matmul(X_T, X)
XTX
```

```python
# Invert XTX
inv_XTX = np.linalg.inv(XTX)
inv_XTX
```

```python
# Create an array y with values [1100, 1300, 800, 900, 1000, 1100, 1200]
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
# Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.
XTX_X = np.matmul(inv_XTX, X_T)
w = np.matmul(XTX_X, y)
w
```

```python
w.sum()
```

The sum of all the elements of w is 0.52.
