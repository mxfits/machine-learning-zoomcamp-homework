## DataFrame

```python
data = [
    ['Nissan', 'Stanza', 1991, 138, 4, 'MANUAL', 'sedan', 2000],
    ['Hyundai', 'Sonata', 2017, None, 4, 'AUTOMATIC', 'Sedan', 27150],
    ['Lotus', 'Elise', 2010, 218, 4, 'MANUAL', 'convertible', 54990],
    ['GMC', 'Acadia',  2017, 194, 4, 'AUTOMATIC', '4dr SUV', 34450],
    ['Nissan', 'Frontier', 2017, 261, 6, 'MANUAL', 'Pickup', 32340],
]

columns = [
    'Make', 'Model', 'Year', 'Engine HP', 'Engine Cylinders',
    'Transmission Type', 'Vehicle_Style', 'MSRP'
]
```

```python
df = pd.DataFrame(data, columns=columns)
df
```

```python
### Let's say we want to add a column 'id'

df['id'] = ['nis1', 'hyu1', 'lot2', 'gmc1', 'nis2']
df
```

```python
### We can change the values of a column
df['id'] = [1, 2, 3, 4, 5]
df
```

```python
### If we want to delete a column

del df['id']
df
```

# Index

```python
df.index
```

```python
df.Make
```

```python
df.Make.index
```

```python
### We can access the row by calling the index
df.loc[2]
```

```python
### Access multiple rows
df.loc[[0, 2, 3]]
```

```python
### Replace index
df.index = ['a', 'b', 'c', 'd', 'e']
df
```

```python
df.loc[['b', 'c']]
```

```python
### locate row using the default index
df.iloc[[0, 1, 2]]
```

```python
### Reset indexing using the default

df.reset_index()  # keeps the new indexing made
```

```python
### If we want to drop the new index made

df.reset_index(drop=True)
```

```python
### Notice that index functions does not change the dataframe

df
```

```python
### To apply changes to the dataframe

df = df.reset_index(drop=True)
df
```
