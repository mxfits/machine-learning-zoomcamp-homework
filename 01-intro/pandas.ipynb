## Grouping

For example, the code below groups the entries by transmission type and calculates the average MSRP for each transmission type.

In SQL:

```
SELECT
    transmission_type,
    AVG(MSRP)
FROM
    cars
GROUP BY
    transmission_type
```

In Python (Pandas):

````python
df.groupby('Transmission Type').MSRP.mean()
````

````python
df.groupby('Transmission Type').MSRP.max()
````

````python
df.groupby('Transmission Type').MSRP.mean()
````
