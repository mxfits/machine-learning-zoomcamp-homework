# Indexing and Slicing Arrays

```python
arr = np.array([[2, 3, 4], [4, 5, 6]])
arr
```

Numpy array slicing

```python
### Get the first row

arr[0]
```

```python
### Get the first column

arr[:,0]  ### Extracts all rows but only the first column
```

Numpy array indexing

```python
### Access the first row, second column

arr[0, 1]
```

```python
### Change the value of an object in an array

arr[1, 2] = 7
```

```python
arr
```

```python
### Change the value of a column

arr[:, 2] = [1, 2]
```

```python
arr
```

# Generating Random Arrays

```python
### Generating an array of random values with 5 rows and 2 columns
### Everytime we execute this, we get a different set of values

np.random.rand(5, 2)
```

```python
np.random.seed(10)
random_array = np.random.rand(5, 2)
random_array
```

```python
### To generate random integers
np.random.seed(10)
random_int = np.random.randint(low=0, high=100, size=(5, 2)) # where 0 is included and 100 is excluded so it generates numbers from 0 to 99
random_int
```

# Array Operations

## Element-wise operations

```python
a = np.arange(5)
a
```

```python
# Additon to each element
a + 1
```

```python
# Subtraction to each element
a - 0.01
```

```python
# Multiplication to each element
a * 2
```

```python
# Division to each element
a / 2
```

```python
# Raising each element by a power
a ** 2
```

```python
# Combination of operations to each element
b = (10 + (a * 2))**2 /100
b
```
