# Converting Lists to Arrays

```python
### let's say we have a list [1, 2, 3, 5, 7, 12]
### To make this an array:

array = np.array([1, 2, 3, 5, 7, 12])
```

```python
array
```

```python
### Another example:

list = [2, 3, 4]
array_from_list = np.array(list)
```

```python
array_from_list
```

```python
### To access an element of an array, like object 4

array_from_list[1]
```

```python
### To change the value of an object in an array:

array_from_list[1] = 10
array_from_list
```

# Indexing and Slicing Arrays

```python
arr = np.array([[2, 3, 4], [4, 5, 6]])
arr
```

Numpy array slicing

```python
### Get the first row

arr[0]
```

```python
### Get the first column

arr[:,0]  ### Extracts all rows but only the first column
```

Numpy array indexing

```python
### Access the first row, second column

arr[0, 1]
```
